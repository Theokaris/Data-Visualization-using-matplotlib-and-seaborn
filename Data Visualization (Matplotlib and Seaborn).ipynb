```python
# importing the required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# importing the file
customers = pd.read_csv("CustomersAssignment.csv")

# Displaying the first five rows
customers.head()
```

```python
# Before doing anything on the data, a copy of original data will be copied for reference
data = customers.copy()
```

```python
# Displaying the first two rows of the copied data
data.head(2)
```

```python
# Checking for duplicates
print('Initial DataFrame shape before duplicates removal:', data.shape)
```

```python
# Removing duplicates if any
data.drop_duplicates(inplace = True)

print('DataFrame shape after duplicates removal:', data.shape)
```

> No duplicates values in the data

```python
# Renaming items in MaritalStatus
data["MaritalStatus"].replace({
    "M":"Married",
    "S":"Single"
}, inplace = True)

data.head(2)
```

```python
# Renaming items in Gender
data["Gender"].replace({
    "M":"Male",
    "F":"Female"
}, inplace = True)

data.head(2)
```

```python
# Renaming items in HomeOwner
data["HomeOwner"].replace({
    "Y":"Yes",
    "N":"No"
}, inplace = True)

data.head(2)
```

```python
# Removing columns not needed for analysis
data.drop(columns = ["Prefix", "Unnamed: 5", "Unnamed: 6", "Unnamed: 8", "Unnamed: 17", "Unnamed: 18","BirthDate.1"], 
          axis = 1, inplace = True)
```

```python
data.head(2)
```

```python
# Renaming Unnamed: 4 to fullname
data.rename(columns={
    "Unnamed: 4":"FullName"
}, inplace = True)

data.head(2)
```

```python
# Checking for null values
data.isnull().sum()
```

> As seen there are 130 null values in Gender column. These will be filled using foward fill

```python
# Filling null values with forward fill
data.fillna(method = "ffill", inplace = True)
```

```python
data.isnull().sum()
```

```python
# Checking datatype of each columns
data.info()
```

> BirthDate column and AnnualIncome columns is of object datatypes rather than datetime and int respectively

```python
# importing the parse function from the dateutil.parser module
# This function can automatically detect and parse various date formats. This will be used to create a function
from dateutil.parser import parse

#creating the clean_date function
def clean_date(text):
    try:
        return parse(text)
    except ValueError:
        return text
```

```python
# Applying the clean_date function to the date column
data["BirthDate"] = data["BirthDate"].apply(clean_date)
data["BirthDate"] = pd.to_datetime(data["BirthDate"])

data["BirthDate"].info()
```

> Date datatype changed to datetime

```python
# Changing the AnnualIncome from object to int
data["AnnualIncome"] # Returns objects from the column
```

```python
# Removing the "$" symbol
data["AnnualIncome"] = data["AnnualIncome"].str.strip("$")
```

```python
# Removing the ","
data["AnnualIncome"] = data["AnnualIncome"].str.replace(",", "")
```

```python
data["AnnualIncome"].sample(n=2) # Returns two random rows
```

```python
# converting from str to int
data["AnnualIncome"] = data["AnnualIncome"].astype(int)
```

```python
data["AnnualIncome"].info()
```

> AnnualIncome column changed from object to int

# Creating a pie chart to represent the proportion of customers by Occupation within a specific income range.

```python
data.AnnualIncome.unique()
```

```python
#Creating a column for Income range
def income_range(x):
    if x <= 50000:
        return "Low Income"
    elif x < 150000:
        return "Average Income"
    else:
        return "High Income"
```

```python
# Applying function to new IncomeRange column
data["IncomeRange"] = data["AnnualIncome"].apply(income_range)

data.head(2)
```

```python
# Grouping by ‘Occupation’ and counting the occurrences of each ‘IncomeRange’ category
grouped_income = data.groupby("Occupation")["IncomeRange"].value_counts()
```

```python
# Creating the pie chart using matplotlib
plt.figure(figsize = (10,8)) # Places the pie chart in this size
grouped_income.plot(kind = "pie", labels = grouped_income.index, autopct = "%1.0f%%", startangle = 90)
plt.title("Pie Chart of Income Range/Occupation Distribution")
plt.ylabel("")
```

> This pie chart provides insights into the distribution of customers based on their income and occupation. The largest segment (26%) represents Professionals with Average Income
